```python
minimum=0
maximum=500000
oddtotal=0
for number in range(minimum, maximum+1):
    if(number%2 !=0):
        print("{0}".format(number))
        oddtotal= oddtotal+number
    print("the sum of odd numbers from {0} to {1})={2}". format(minimum, maximum, oddtotal))
```
